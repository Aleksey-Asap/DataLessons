```python
def transform_country(country):
    if country == "Spain": 
        return "2023"
    elif country == "Germany":
        return "2024"
    else: 
        return ""
    
transform_country("Spain")
```

```python
## Самый тупой способ\
for value in df["Country"]:
    print(transform_country(value))
```

```python
## Самый тупой способ\
values = []
for value in df["Country"]:
    values.append(transform_country(value))
values
```

```python
df["car_year"] = values 
df
```

```python
## Как надо:
df["car_year"] = df["Country"].apply(transform_country)
df
```
