### Объяснение ООП - Объектно Ориентированное Программирование (Класс и Объект)

```python
class Car:
    def __init__(self, make, model, year, color, n_doors):
        """"
        Метод инициализирующий свойсва обектов класса 
    
        """
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.n_doors = n_doors
        self.current_speed = 0


    def get_info(self):
        """
        Выводит информацию о свойствах объекта класса
        """
        car_info = f"Производитель: {self.make} Модель: {self.model} Год Выпуска: {self.year} Цвет: {self.color} Число Дверей: {self.n_doors}"
        return car_info.title()


    def get_speed(self):
        """
        Получить текущую скорость авто 
        """
        print(f"Текущая скорость {self.current_speed} км/ч")

    
    def updte_speed(self, inc):
        """
        Обновить скорость авто
        """
        self.current_speed += inc


    def stop_car(self):
        self.current_speed = 0
        print('Автомобиль остановлен')

    
    def stop_engine(self):
        print('Двигатель был выключен')


    def start_engine(self):
        print('Двигатель заводится')
        print('Двигатель работает. Можно ехать')
```

```python
car_1 = Car(make='audi', model='a4', year='2000', color='red', n_doors=2)
car_2 = Car(make='lada', model='granta', year='2015', color='blue', n_doors=5)
car_3 = Car(make='bmw', model='3', year='2020', color='black', n_doors=5)
```

```python
car_1.get_info()
```

```python
car_3.get_info()
```

```python
car_1.get_speed()
```

```python
# имитируем движение авто
lst_speed = [30, 100, 250]

print(car_1.get_info())
print()

for speed_inc in lst_speed:
    car_1.updte_speed(speed_inc)
    car_1.get_speed()

car_1.stop_car()
car_1.stop_engine()

print()
car_1.get_speed()
```

```python
# создаем переменную куда записываем записи продаж по жанру Action
action_genre_sales = games_data[games_data['Genre'] == 'Action']
```

```python
# группировка -> определить динамику продаж по годам
action_sales_agg = (
    action_genre_sales
        .groupby('Year', as_index=True)
        .agg({
            'North America': 'sum',
            'Europe': 'sum', # SUM(COLUN)
            'Japan': 'sum',
            'Rest of World': 'sum'
        })
)

action_sales_agg
```

```python
# рисуем
plt.figure(figsize=(12, 6))
action_sales_agg['North America'].plot(label='North America')
action_sales_agg['Europe'].plot(label='Europe')
action_sales_agg['Japan'].plot(label='Japan')
action_sales_agg['Rest of World'].plot(label='Rest of World')

plt.title(f'Action Sales')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# пример -> имеем 2 DataFrame, как их соеденить
columns_1 = [
    'Publisher',
    'North America'
]

columns_2 = [
    'Publisher',
    'Europe'
]

df_1 = games_data[columns_1].iloc[:10]
df_2 = games_data[columns_2].iloc[:10]

# мерджим 2 df
merged_df = df_1.merge(df_2, on='Publisher', how='left')
merged_df
```

```python
# почему имеем дубликаты?
df_1_publishers = set(df_1['Publisher'])
df_2_publishers = set(df_2['Publisher'])

intersected_columns = df_1_publishers.intersection(df_2_publishers)
print(intersected_columns)
```

```python
# проверим уникальность строк 
merged_df.drop_duplicates(subset=['Publisher', 'North America', 'Europe'])
```
